```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

Tenemos el siguiente sistema de ecuación diferenciales de segundo grado acopladas que modela el movimiento de la carga $q$ con masa $m$ dentro de un campo magnético $B_o$ homogéneo.

$$
\frac{d^x}{dt^2} = 0 \\
\frac{d^2y}{dt^2} = w_o\frac{dz}{dt}\\
\frac{d^2z}{dt^2} = -w_o\frac{dy}{dt}\\
$$

En donde $w_0 = \frac{q}{m}B_o$.

```python

ax = plt.figure().add_subplot(projection='3d')

# Make the grid
x, y, z = np.meshgrid(np.arange( 0, 8, 2),np.arange( 0, 8, 2),np.arange( 0 ,8, 2))

u = 1
v = 0
w = 0
ax.quiver(x, y, z, u, v, w, length= 0.8, normalize=True, label='Campo Magnetico')
ax.scatter(3,3,3, color='red', label='Particula')
plt.legend()
plt.savefig('casoinicial.pdf')
```

```python
#Definimos sistema de ecuaciones 

def ecmov(p,t,wo):
    x , y , z, dx , dy , dz = p
    dp = (dx , dy , dz , 0 , wo*dz, -wo*dy)
    return dp
```

```python
#Definir variables

t = np.linspace(0,10,1000)
wo = 2*np.pi/2
```

```python
#Caso dx =/ 0, dy = dz = 0 

p1 = (0,0,0,1,0,0)

sol1 = odeint(ecmov ,p1 , t , args=(wo,))
x1 = sol1[:,0]
y1 = sol1[:,1]
z1 = sol1[:,2]
```

```python
ax = plt.figure(figsize=(8,6)).add_subplot(projection='3d')

ax.plot(x1, y1, z1, color='red', lw=1)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("")

plt.savefig('caso1.pdf')
```

```python
#Caso dx = 0, dy =/ dz =/ 0 

p2 = (0,0,0,0,1,1)

sol2 = odeint(ecmov ,p2 , t , args=(wo,))
x2 = sol2[:,0]
y2 = sol2[:,1]
z2 = sol2[:,2]
```
